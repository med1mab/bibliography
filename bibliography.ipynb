```python
def View_search():
    import xml.etree.ElementTree as ET
    import pandas as pd
    mytree = ET.parse('BD.xml')
    root = mytree.getroot()
    i = 1
    datag = []
    data = []
    for s in root.findall('./dblp/inproceedings'):
        ch = "authors  :"
        for i, child in enumerate(s):
            if(child.tag == 'author'):
                ch = ch+","+child.text
                continue
            if(child.tag == 'ee' or child.tag == 'crossref' or child.tag == 'url'):
                continue
            c = child.tag+"  :"+child.text
            data.append(c)
        data.append(ch)
        datag.append(data)
        for l in data:
            print(l)
        data = []
        print()
    ch = ""
    for l in datag:
        c = ""
        for t in l:
            c = c+"\n"+t
        ch = ch+"\n"+c
    Num_Vertical = (ch)

    window = tk.Tk()
    window.title("View_search")
    window.minsize(width=400, height=400)
    window.geometry("600x500")
    window.config(bg='pink')
    SVBar = tk.Scrollbar(window)
    SVBar.pack(side=tk.RIGHT,
               fill="y")

    SHBar = tk.Scrollbar(window,
                         orient=tk.HORIZONTAL)
    SHBar.pack(side=tk.BOTTOM,
               fill="x")

    TBox = tk.Text(window,
                   height=500,
                   width=500,
                   yscrollcommand=SVBar.set,
                   xscrollcommand=SHBar.set,
                   wrap="none")

    TBox = tk.Text(window,
                   height=500,
                   width=500,
                   yscrollcommand=SVBar.set,
                   xscrollcommand=SHBar.set,
                   wrap="none")

    TBox.pack(expand=0, fill=tk.BOTH)

    TBox.insert(tk.END, Num_Vertical)

    SHBar.config(command=TBox.xview)
    SVBar.config(command=TBox.yview)

    window.mainloop()
```
